# Problem Statement

Solve the Cylinder Problem considering the following data:

* N: 10
* $c_1$: 2
* $c_2$: 0.5
  

```julia
# Minimize surface area for a given volume

# 1. Definição do Modelo de Otimização
model = Model(Ipopt.Optimizer)

# Parâmetros baseados no documento [cite: 80]
N = 10.0      
c1 = 2.0      
c2 = 0.5      

@variable(model, r >= 0, start = 1.0)
@variable(model, h >= 0, start = 1.0)

# Função objetivo definida no problema [cite: 52, 65]
@NLobjective(model, Max, 
    N * (π * r^2 * h) - 
    c1 * (π * r^2) - 
    (c2 * π * r^2 * h) * (π * r^2 + 2 * π * r * h)
)

optimize!(model)

r_opt = value(r)
h_opt = value(h)
z_opt = objective_value(model)

# 2. Exibição dos Resultados
# 
println("--- Resultado da Otimização ---")
println("Raio ótimo (r): ", value(r))
println("Altura ótima (h): ", value(h))
println("Lucro Máximo (z): ", objective_value(model))

# 3. Preparação dos Gráficos
# Definindo os intervalos para visualização similares ao MATLAB no arquivo [cite: 70]
r_range = 0.1:0.05:2.5
h_range = 0.1:0.05:3.5

# Função de lucro para o plot
f(r, h) = N*(π*r^2*h) - c1*(π*r^2) - (c2*π*r^2*h)*(π*r^2 + 2*π*r*h)

# 4. Gerando o Gráfico de Superfície 3D [cite: 75, 78]
p1 = surface(r_range, h_range, f, 
    title="Superfície de Lucro",
    xlabel="Raio (r)", ylabel="Altura (h)", zlabel="Lucro (z)",
    color=:viridis, camera=(30, 30))
scatter!(p1, [r_opt], [h_opt], [z_opt], color=:red, markersize=5, label="Ótimo")

# 5. Gerando o Gráfico de Contorno (Visualização de Topo)
p2 = contour(r_range, h_range, f, 
    title="Curvas de Nível",
    xlabel="Raio (r)", ylabel="Altura (h)",
    fill=true, color=:viridis)
scatter!(p2, [r_opt], [h_opt], color=:red, markersize=5, label="Ótimo (r=$r_opt, h=$h_opt)")

# Exibir os gráficos lado a lado
plot(p1, p2, layout=(1, 2), size=(1000, 450))
```

# Problem Statement

Solve the Awning Problem considering the following data:

* h: 2
* w: 3
* initial guess $(x,y) = (1.0, 1.0)$
  

```julia

# 1. Inicialização do Modelo
# Utilizamos o Ipopt por ser um solver robusto para Programação Não Linear (NLP)
model = Model(Ipopt.Optimizer)

# 2. Parâmetros 
# h: altura da caixa, w: largura da caixa
h = 2.0
w = 3.0

# 3. Variáveis de Decisão
# x: distância da parede ao ponto de fixação no solo
# y: altura da fixação na parede
# Ambas devem ser positivas 
@variable(model, x >= 0, start = 1.0)
@variable(model, y >= 0, start = 1.0)

# 4. Função Objetivo 
# Minimizar z = sqrt(x^2 + y^2) - Comprimento do toldo (Hipotenusa)
@NLobjective(model, Min, sqrt(x^2 + y^2))

# 5. Restrição de Cobertura 
# A altura do toldo no ponto w deve ser pelo menos h: y - (y/x)*w >= h
@NLconstraint(model, cobertura, y - (y/x) * w >= h)

# 6. Solução do Problema
optimize!(model)

# 7. Extração dos Resultados
x_opt = value(x)
y_opt = value(y)
z_opt = objective_value(model)

println("--- Resultado da Otimização do Toldo ---")
println("Distância no solo (x): ", round(x_opt, digits=3), " m")
println("Altura na parede (y): ", round(y_opt, digits=3), " m")
println("Comprimento mínimo do toldo (z): ", round(z_opt, digits=3), " m")

# 8. Gráficos de Visualização
# Intervalos para o gráfico
x_range = 3.1:0.1:10.0  # x deve ser > w
y_range = 4.1:0.1:10.0  # y deve ser > h

# Função do comprimento para o plot
f_comprimento(x, y) = sqrt(x^2 + y^2)

# Gráfico de Contorno (Curvas de Nível)
p1 = contour(x_range, y_range, f_comprimento, 
    title="Otimização do Comprimento",
    xlabel="x (Solo)", ylabel="y (Parede)",
    fill=true, color=:viridis)

# Adicionando a linha da restrição de cobertura: y = (h*x) / (x - w)
x_cons = 3.1:0.05:10.0
y_cons = [(h * val) / (val - w) for val in x_cons]
plot!(p1, x_cons, y_cons, color=:white, lw=3, label="Fronteira de Cobertura", ylims=(0, 15))

# Ponto ótimo
scatter!(p1, [x_opt], [y_opt], color=:red, markersize=6, label="Ponto Ótimo (z=$z_opt)")

# Gráfico do Perfil do Toldo (Visualização Geométrica)
p2 = plot([0, x_opt], [y_opt, 0], lw=4, label="Toldo", color=:blue, aspect_ratio=:equal)
plot!([0, w, w, 0, 0], [0, 0, h, h, 0], fill=(0, :grey), label="Caixa (h=4, w=3)")
title!(p2, "Geometria da Solução")
xlabel!(p2, "Largura (x)")
ylabel!(p2, "Altura (y)")

plot(p1, p2, layout=(1, 2), size=(1000, 450))
```

# Problem Statement

Solve the Packing Problem

```julia

# 1. Inicialização do Modelo
# O solver Ipopt é ideal para este problema de maximização de volume (não linear)
model = Model(Ipopt.Optimizer)

# 2. Variáveis de Decisão 
# h: altura, w: largura, d: profundidade (em cm)
@variable(model, h >= 0, start = 1.0)
@variable(model, w >= 0, start = 1.0)
@variable(model, d >= 0, start = 1.0)

# 3. Função Objetivo 
# Maximizar o volume da caixa: V = h * w * d
@NLobjective(model, Max, h * w * d)

# 4. Restrição de Material/Papelão 
# A soma das áreas das faces, considerando peso triplo (3x) no topo e fundo:
# Laterais: 2*w*h + 2*d*h
# Topo e Fundo: 3 * (2*w*d) = 6*w*d
# Limite total: 60 cm²
@NLconstraint(model, papelao, 2*w*h + 2*d*h + 6*w*d <= 60)

# 5. Execução da Otimização
optimize!(model)

# 6. Extração dos Resultados Ótimos
h_opt = value(h)
w_opt = value(w)
d_opt = value(d)
z_opt = objective_value(model)

println("--- Resultado da Otimização da Caixa ---")
println("Altura (h): ", round(h_opt, digits=4), " cm")
println("Largura (w): ", round(w_opt, digits=4), " cm")
println("Profundidade (d): ", round(d_opt, digits=4), " cm")
println("Volume Máximo (z): ", round(z_opt, digits=4), " cm³")

# 7. Visualização Gráfica
# Vamos visualizar como o Volume (z) se comporta variando Largura e Profundidade,
# mantendo a Altura (h) fixa no valor ótimo encontrado.

w_range = 0.5:0.1:5.0
d_range = 0.5:0.1:5.0

p1 = surface(w_range, d_range, (w, d) -> h_opt * w * d,
    title="Superfície de Volume (h fixo)",
    xlabel="Largura (w)", ylabel="Profundidade (d)", zlabel="Volume",
    color=:viridis)

# Adicionando o ponto ótimo no gráfico
scatter!(p1, [w_opt], [d_opt], [z_opt], color=:red, markersize=5, label="Ponto Ótimo")

display(p1)
```

# Problem Statement

Solve the 3-bus Optimal Power Flow Problem with following data:



```julia
# 1. Configuração do Modelo
model = Model(Ipopt.Optimizer)
set_optimizer_attribute(model, "tol", 1e-12)
set_optimizer_attribute(model, "max_iter", 1000)

# 2. Parâmetros 
c1, c2 = 2.0, 1.0           
p_load, q_load = 0.8, 0.6   

# Admitância da linha: y = 1 / (0.01 + j0.1)
# yl ≈ 9.95037190, al ≈ -1.47062891
yl = 9.95037190
al = -1.47062891

# 3. Variáveis de Decisão
@variable(model, 0.95 <= v1 <= 1.10, start = 1.0)
@variable(model, 0.95 <= v2 <= 1.10, start = 1.1)
@variable(model, 0.95 <= v3 <= 1.10, start = 1.0)

# Chutes iniciais para ângulos (Radianos) evitam a solução nula (zero)
@variable(model, d1, start = 0.02)
@variable(model, d2, start = 0.04)
@variable(model, d3 == 0.0) 

@variable(model, 0.0 <= p1 <= 3.0, start = 0.1)
@variable(model, 0.0 <= p2 <= 0.8, start = 0.7)
@variable(model, -2.0 <= q1 <= 2.0, start = 0.3)
@variable(model, -2.0 <= q2 <= 2.0, start = 0.3)

# 4. Função Objetivo 
@objective(model, Min, c1 * p1 + c2 * p2)

# 5. Equações de Balanço de Potência AC 
# Cada barra i tem conexões com as outras duas barras j e k
# Ex: p1 = Fluxo(1->2) + Fluxo(1->3)

# Barra 1 (Conectada a 2 e 3)
@NLconstraint(model, bp1, p1 == 2 * v1^2 * yl * cos(al) - v1 * v2 * yl * cos(d1 - d2 - al) - v1 * v3 * yl * cos(d1 - d3 - al))
@NLconstraint(model, bq1, q1 == -2 * v1^2 * yl * sin(al) - v1 * v2 * yl * sin(d1 - d2 - al) - v1 * v3 * yl * sin(d1 - d3 - al))

# Barra 2 (Conectada a 1 e 3)
@NLconstraint(model, bp2, p2 == 2 * v2^2 * yl * cos(al) - v2 * v1 * yl * cos(d2 - d1 - al) - v2 * v3 * yl * cos(d2 - d3 - al))
@NLconstraint(model, bq2, q2 == -2 * v2^2 * yl * sin(al) - v2 * v1 * yl * sin(d2 - d1 - al) - v2 * v3 * yl * sin(d2 - d3 - al))

# Barra 3 (Carga, conectada a 1 e 2)
@NLconstraint(model, bp3, -p_load == 2 * v3^2 * yl * cos(al) - v3 * v1 * yl * cos(d3 - d1 - al) - v3 * v2 * yl * cos(d3 - d2 - al))
@NLconstraint(model, bq3, -q_load == -2 * v3^2 * yl * sin(al) - v3 * v1 * yl * sin(d3 - d1 - al) - v3 * v2 * yl * sin(d3 - d2 - al))

# 6. Limites de Fluxo Aparente nas Linhas 
# S12 <= 0.25
@NLconstraint(model, l12, sqrt((yl*cos(al)*v1^2 - yl*v1*v2*cos(d1-d2-al))^2 + (-yl*sin(al)*v1^2 - yl*v1*v2*sin(d1-d2-al))^2) <= 0.25)
@NLconstraint(model, l21, sqrt((yl*cos(al)*v2^2 - yl*v2*v1*cos(d2-d1-al))^2 + (-yl*sin(al)*v2^2 - yl*v2*v1*sin(d2-d1-al))^2) <= 0.25)

# S13 e S23 <= 2.0
@NLconstraint(model, l13, sqrt((yl*cos(al)*v1^2 - yl*v1*v3*cos(d1-d3-al))^2 + (-yl*sin(al)*v1^2 - yl*v1*v3*sin(d1-d3-al))^2) <= 2.0)
@NLconstraint(model, l23, sqrt((yl*cos(al)*v2^2 - yl*v2*v3*cos(d2-d3-al))^2 + (-yl*sin(al)*v2^2 - yl*v2*v3*sin(d2-d3-al))^2) <= 2.0)

# 7. Execução
optimize!(model)

# 8. Saída de Dados 
println("\n" * "="^40)
println("   RESULTADOS FINAIS - OPF 3 BUS")
println("="^40)
@printf("p1.L  = %.8f\n", value(p1))
@printf("p2.L  = %.8f\n", value(p2))
@printf("q1.L  = %.8f\n", value(q1))
@printf("q2.L  = %.8f\n", value(q2))
@printf("v1.L  = %.8f\n", value(v1))
@printf("v2.L  = %.8f\n", value(v2))
@printf("v3.L  = %.8f\n", value(v3))
@printf("d1.L  = %.8f\n", value(d1))
@printf("d2.L  = %.8f\n", value(d2))
println("="^40)# code
```

# Problem Statement

A small engineering consulting firm has 3 senior designers available to work on the firm's 4 current projects over the next 2 weeks. Each designer has 80 hours to split among the projects, and the following table shows the manager's scoring $(0=$ nil to $100=$ perfect $)$ of the capability of each designer to contribute to each project, along with his estimate of the hours that each project will require.


|      Designer     | Project 1    | Project 2    | Project 3    | Project 4    |
|----------|--------------|--------------|--------------|--------------|
| 1        | 90           | 80           | 10           | 50           |
| 2        | 60           | 70           | 50           | 65           |
| 3        | 70           | 40           | 80           | 85           |



|     **Required:**      | Project 1    | Project 2    | Project 3    | Project 4    |
|-----------|--------------|--------------|--------------|--------------|
| **Hours** | 70           | 50           | 85           | 35           |


## Model

Let the design engineers be set $E$ with $E_{i} \; :i \in [1,2,3]$ and the projects be $P$ with  $P_{j} \; :j \in [1,2,3,4]$. We can model the problem as allocation of the number of hours $H_{ij}$ with each design engineer $E_{i}$ $\forall i $ that are being put onto the projects $P_{j}$ $\forall j $, given the $i^{th}$ engineer $E$ works on $j^{th}$ project with given score $e_{ij}$.

Let the maximum hours available with each engineer be $H_{max}$ and the required number of hours for each project $P_j$ be $R_{j} \; \forall j $. Thus, then the mathematical formulation can be made as:

$$
\text{Maximize } \sum_{i \in E} \sum_{j \in P} H_{ij} e_{ij}
$$

Subject to:

$$
\sum_{j \in P} H_{ij} \leq H_{max} \;\; \forall i \in E
$$

$$
\sum_{i \in E} H_{ij} \geq R_j \;\; \forall j \in P
$$

$$
H_{ij} \geq 0 \;\; \forall i,j
$$


## Implement

```julia
# Data
H_max = 80
scores = [
    90  80  10  50;
    60  70  50  65;
    70  40  80  85
]
required = [70, 50, 85, 35]
n_designers = 3
n_projects = 4

# Solve optimization
model = Model(GLPK.Optimizer)
@variable(model, H[1:n_designers, 1:n_projects] >= 0)
@objective(model, Max, sum(H[i, j] * scores[i, j] for i in 1:n_designers, j in 1:n_projects))
for i in 1:n_designers
    @constraint(model, sum(H[i, j] for j in 1:n_projects) <= H_max)
end
for j in 1:n_projects
    @constraint(model, sum(H[i, j] for i in 1:n_designers) >= required[j])
end
optimize!(model)

H_opt = value.(H)
allocated = [sum(H_opt[i, j] for i in 1:n_designers) for j in 1:n_projects]

# Display results
println("Optimal Allocation:")
println("Designer | P1   P2   P3   P4  | Total")
println("-"^40)
for i in 1:n_designers
    println("D$i       ", join([round(H_opt[i, j], digits=1) for j in 1:n_projects], "  "), " | ", round(sum(H_opt[i, :]), digits=1))
end

println("\nProject Totals:")
println("Project | Required | Allocated | Status")
println("-"^40)
for j in 1:n_projects
    status = allocated[j] >= required[j] ? "✓ Met" : "✗ Under"
    println("P$j       | ", required[j], "       | ", round(allocated[j], digits=1), "       | $status")
end

println("\nObjective value: ", round(objective_value(model), digits=2))

# Simple plots (guaranteed to work)
# Heatmap
heatmap(H_opt, 
        title="Hours Allocation Heatmap",
        xlabel="Projects", 
        ylabel="Designers",
        xticks=(1:4, ["P1", "P2", "P3", "P4"]),
        yticks=(1:3, ["D1", "D2", "D3"]))

# Requirements vs Allocated side by side using subplot
p1 = bar(1:4, required, label="Required", color=:blue, title="Project Hours", ylabel="Hours")
p2 = bar(1:4, allocated, label="Allocated", color=:orange, title="Project Hours", ylabel="Hours")
plot(p1, p2, layout=(1, 2), size=(800, 400))

# Designer utilization
bar(1:3, [sum(H_opt[i, :]) for i in 1:3], 
    label="Hours Used", 
    title="Designer Utilization",
    xlabel="Designer",
    ylabel="Hours",
    xticks=(1:3, ["D1", "D2", "D3"]),
    color=:green)
hline!([80], label="Capacity (80h)", color=:red, linestyle=:dash)
```

```julia
# Sensitivity Analysis

println("\n" * "="^70)
println("SENSITIVITY ANALYSIS")
println("="^70)

# Check which constraints are binding
println("\nBinding Constraints:")
println("-"^50)

# Designer capacity constraints
for i in 1:n_designers
    total = sum(value(H[i, j]) for j in 1:n_projects)
    if total >= H_max - 1e-6
        println("✓ Designer $i is fully utilized ($(round(total, digits=1))/80 hours)")
    else
        println("  Designer $i has slack: $(round(H_max - total, digits=1)) hours remaining")
    end
end

println()
# Project requirement constraints (dual/shadow prices)
for j in 1:n_projects
    allocated = sum(value(H[i, j]) for i in 1:n_designers)
    slack = allocated - required[j]
    if slack <= 1e-6
        println("✓ Project $j requirement is exactly met")
    else
        println("  Project $j has excess: $(round(slack, digits=1)) hours above requirement")
    end
end

# Efficiency analysis
println("\n" * "-"^70)
println("EFFICIENCY ANALYSIS (Score per hour)")
println("-"^70)

for i in 1:n_designers
    println("\nDesigner $i:")
    for j in 1:n_projects
        if H_opt[i, j] > 0.1
            efficiency = scores[i, j]
            println("  Project $j: $(H_opt[i, j]) hours @ score $efficiency")
        end
    end
end
```

# Problem Statement

A dietitian is planning a meal that meets the daily nutritional requirements for calories, protein, and vitamins at a minimum cost.


|      Food Item     | Cost ($)    | Calories    | Protein (g)    | Vitamins (% Daily)    |
|----------|--------------|--------------|--------------|--------------|
| Apple        | 1              | 100           | 0.5           | 2            |
| Bread        | 0.50           | 200           | 4             | 0            |
| Milk         | 2              | 150           | 8             | 10           |
| Egg          | 0.30           | 70            | 6             | 0            |

Daily nutritional requirements: 500 calories, 50g protein, 100% vitamins.

Define decision variables: $y_1$ for Apples, $y_2$ for Bread, $y_3$ for Milk, $y_4$ for Eggs.  
$y_i$ represents the quantity of each food item.

$$
\begin{aligned}
\text{Minimize} \quad & y_1 + 0.5y_2 + 2y_3 + 0.3y_4 \\
\text{Subject to} \quad 
& 100y_1 + 200y_2 + 150y_3 + 70y_4 \geq 500 \\
& 0.5y_1 + 4y_2 + 8y_3 + 6y_4 \geq 50 \\
& 2y_1 + 0y_2 + 10y_3 + 0y_4 \geq 100 \\
& y_1, y_2, y_3, y_4 \geq 0
\end{aligned}
$$

Ensure all dietary requirements for calories, protein, and vitamins are met.

```julia
# Data
cost = [1.0, 0.5, 2.0, 0.3]  # $ per unit
calories = [100, 200, 150, 70]  # calories per unit
protein = [0.5, 4.0, 8.0, 6.0]  # grams per unit
vitamins = [2, 0, 10, 0]  # % daily value per unit

# Requirements
min_calories = 500
min_protein = 50
min_vitamins = 100

n_foods = 4

# Create optimization model
model = Model(GLPK.Optimizer)

# Decision variables: quantity of each food
@variable(model, y[1:n_foods] >= 0)

# Objective: Minimize total cost
@objective(model, Min, sum(cost[i] * y[i] for i in 1:n_foods))

# Constraints
@constraint(model, sum(calories[i] * y[i] for i in 1:n_foods) >= min_calories)
@constraint(model, sum(protein[i] * y[i] for i in 1:n_foods) >= min_protein)
@constraint(model, sum(vitamins[i] * y[i] for i in 1:n_foods) >= min_vitamins)

# Solve
optimize!(model)

# Display results
println("="^60)
println("DIET OPTIMIZATION SOLUTION")
println("="^60)

println("\nSolution status: ", termination_status(model))
println("Minimum total cost: \$", round(objective_value(model), digits=2))

println("\n" * "-"^60)
println("OPTIMAL FOOD QUANTITIES")
println("-"^60)

food_names = ["Apple", "Bread", "Milk", "Egg"]
for i in 1:n_foods
    quantity = value(y[i])
    if quantity > 0.001
        println("$i. $(food_names[i]): $(round(quantity, digits=2)) units (\$$(round(quantity * cost[i], digits=2)))")
    else
        println("$i. $(food_names[i]): 0 units (not selected)")
    end
end

println("\n" * "-"^60)
println("NUTRITIONAL ANALYSIS")
println("-"^60)

# Calculate actual nutritional values
total_calories = sum(calories[i] * value(y[i]) for i in 1:n_foods)
total_protein = sum(protein[i] * value(y[i]) for i in 1:n_foods)
total_vitamins = sum(vitamins[i] * value(y[i]) for i in 1:n_foods)

println("Calories:  $(round(total_calories, digits=1)) / $min_calories (≥ requirement)")
println("Protein:   $(round(total_protein, digits=1)) g / $min_protein g (≥ requirement)")
println("Vitamins:  $(round(total_vitamins, digits=1))% / $min_vitamins% (≥ requirement)")

println("\n" * "-"^60)
println("COST BREAKDOWN")
println("-"^60)

for i in 1:n_foods
    quantity = value(y[i])
    if quantity > 0.001
        println("$(food_names[i]): $(round(quantity, digits=2)) units × \$$(cost[i]) = \$$(round(quantity * cost[i], digits=2))")
    end
end
println("Total: \$$(round(objective_value(model), digits=2))")
```

# Problem Statement

Consider a hiker who needs to choose the most valuable items for a hike without overloading the backpack.

- Items: Tent (Value: $\$120$, Weight: 2kg), Stove (Value: $\$80$, Weight: 1kg), Food (Value: $\$60$, Weight: 1kg)  
- Backpack capacity: 3.5kg  

Objective: Maximize the value of items in the backpack.

Define binary decision variables: $x_1$ for Tent, $x_2$ for Stove, $x_3$ for Food.  
$x_i = 1$ if the item is chosen, and 0 otherwise.

$$
\begin{aligned}
\text{Maximize} \quad & 120x_1 + 80x_2 + 60x_3 \\
\text{Subject to} \quad 
& 2x_1 + x_2 + x_3 \leq 3.5 \\
& x_1, x_2, x_3 \in \{0,1\}
\end{aligned}
$$

```julia
# Using HiGHS solver (often faster for integer problems)

# Data (defined locally to avoid missing variables)
values = [120, 80, 60]
weights = [2.0, 1.0, 1.0]
capacity = 3.5
item_names = ["Tent", "Stove", "Food"]
n_items = 3

# Using HiGHS solver
model_highs = Model(HiGHS.Optimizer)

@variable(model_highs, x[1:n_items], Bin)
@objective(model_highs, Max, 120x[1] + 80x[2] + 60x[3])
@constraint(model_highs, 2x[1] + x[2] + x[3] <= 3.5)

optimize!(model_highs)

println("\n" * "="^60)
println("SOLUTION WITH HIGHS SOLVER")
println("="^60)
println("Maximum value: \$", objective_value(model_highs))
println("\nSelected items:")
for i in 1:n_items
    if value(x[i]) > 0.5
        println("  ✓ $(item_names[i])")
    else
        println("  ✗ $(item_names[i])")
    end
end

# Additional details
println("\n" * "-"^60)
println("DETAILED SOLUTION:")
println("-"^60)
total_weight = 0.0
for i in 1:n_items
    if value(x[i]) > 0.5
        total_weight += weights[i]
        println("  $(item_names[i]): $(weights[i]) kg, value = \$$(values[i])")
    end
end
println("  Total weight: $(total_weight) kg / $capacity kg")
println("  Remaining capacity: $(capacity - total_weight) kg")
```
